```python
with open("data.txt","r") as f:
    data = [int(i.strip()) for i in f.readlines()]
```

```python
previous = 0
count = -1
for num in data:
    if num > previous:
        print("increased", num)
        count +=1
        previous = num
    else:
        print("decreased", num)
        previous =  num
print(count)
    
```

```python
#Part Two

previous = 0
count = -1

for i in range(0,len(data)-2):
    window_sum = 0
    for window in data[i:i+3]:
        window_sum += window
    if window_sum > previous:
        print("increased", window_sum, previous)
        count +=1
        previous = window_sum
    else:
        print("decreased", window_sum)
        previous =  window_sum
print(count)
```
