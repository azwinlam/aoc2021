```python
with open("data06.txt","r") as f:
    data = [i.strip().replace("->","").split() for i in f.readlines()]
input = data[0][0].split(',')
def lanterFishModel(x):
    if x == 0:
        return 6
    else:
        return x-1


input = [int(i) for i in input]
dayCount = 1
new_sequence = []

def dayPassing(sequence):
    new_sequence = []
    for i in sequence:
        result = lanterFishModel(i)
        new_sequence.append(result)
    # print(new_sequence)
    return new_sequence


while dayCount < 100:
    new_input = dayPassing(input)
    dayCount += 1
    for j in new_input:
        if j == 0:
            new_input.append(9)
    input = new_input
    print(len(input))
```
